# Data Cleaning on Retail Sales Dataset



## Objective

The objective of this project is to clean the Vrinda Store retail sales dataset by identifying and handling missing values, duplicate records, incorrect data types, inconsistent formatting, and invalid values. A clean dataset improves the quality of analysis and helps generate reliable business insights.

The dataset was checked for duplicate records. No duplicate rows were found after the cleaning process.

```python
# Remove extra spaces from Column names
df.columns = df.columns.str.strip()
```

```python
# Convert date column 
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df['Date'].dtype
```

```python
# Check negative revenue
df[df['Amount'] < 0]
```

```python
# Check invalid quantity

# print(df['Qty'].value_counts())
df['Qty'] = df['Qty'].replace({"One" : 1,
                              "Two" : 2})
df['Qty']=pd.to_numeric(df['Qty'])

df[df['Qty'] < 0]
```

```python
df['Gender'] = df['Gender'].str.title()

#standardize the values in the **Gender** column by replacing different representations of the same gender with a consistent format.

df['Gender'] = df['Gender'].replace({
    'M': 'Men',
    'm': 'Men',
    'Man': 'Men',
    'W': 'Women',
    'w': 'Women',
    'Woman': 'Women'
})
```

```python
df['Gender'].value_counts()
```

```python
#Standardize Category
df['Category'] = df['Category'].str.title()
```

```python
#Standardize Channel
df['Channel'] = df['Channel'].str.title()
```

```python
#Standardize state
df['ship-state'] = df['ship-state'].str.title()
```

```python
# Check outliers
df[['Age','Qty','Amount']].describe()
```

```python
# Final missing value check
df.isnull().sum()
```

```python
# Final duplicate value check
df.duplicated().sum()
```

```python
# Final shape
df.shape
```

The final dataset contains all valid records and is ready for further analysis and visualization.
